```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# 使用np.random.RandomState功能创建随机数据集
rng=np.random.RandomState(1)
print(rng)
rng.rand(80,1)
X=np.sort(5*rng.rand(80,1),axis=0)
y=np.sin(X).ravel()
y
```

```python
len(y[::5])
```

```python
y[::5]+=3*(0.6-rng.rand(16))
```

```python
y
```

```python
# 决策树回归模型
regr_1=DecisionTreeRegressor(max_depth=2)
regr_2=DecisionTreeRegressor(max_depth=6)
```

```python
regr_1.fit(X,y)
```

```python
regr_2.fit(X,y)
```

```python
# 预测
X_test=np.arange(0,5,0.01)[:,np.newaxis]
X_test
```

```python
y_1=regr_1.predict(X_test)
y_2=regr_2.predict(X_test)
```

```python
y_1
```

```python
y_2
```

```python
# 汉字设置
mpl.rcParams['font.sans-serif'] = ['simHei']
mpl.rcParams['axes.unicode_minus'] = False
# 绘图
plt.figure()
plt.scatter(X,y,s=20,edgecolor="black",label='data',c='darkorange')
plt.plot(X_test,y_1,color='cornflowerblue',label='最大深度=2',linewidth=2)
plt.plot(X_test,y_2,color='yellowgreen',label='最大深度=5',linewidth=2)
plt.xlabel('随机自变量')
plt.ylabel('因变量')
plt.title('决策树实现回归任务示例')
plt.show()
```
