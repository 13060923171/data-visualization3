```python
# 导入需要的模块
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import scipy.stats as sci
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
plt.style.use('ggplot')
```

```python
data = pd.read_csv('../input/train.csv', index_col=0)
```

```python
# 查看数据特征类型, 即除SalePrice的列，有数值类型(numerical)和类别类型（category）
numeric = [f for f in data.drop(['SalePrice'], axis=1).columns if data.drop(['SalePrice'], axis=1).dtypes[f] != 'object']
category = [f for f in data.drop(['SalePrice'], axis=1).columns if data.drop(['SalePrice'], axis=1).dtypes[f] == 'object']
print('numeric:')
print(numeric)
print('category:')
print(category)
print("numeric: {}, category: {}" .format (len(numeric),len(category)))
```

```python
# 缺失值处理
# 缺失值统计
missing = data.isnull().sum()
missing.sort_values(inplace=True,ascending=False)
missing = missing[missing > 0]
types = data[missing.index].dtypes
percent = (data[missing.index].isnull().sum()/data[missing.index].isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([missing, percent,types], axis=1, keys=['Total', 'Percent','Types'])
missing_data.sort_values('Total',ascending=False,inplace=True)
missing_data
```

```python
missing.plot.bar()
```

```python
# 对缺失值超过15%的特征进行删除
data.drop(missing_data[missing_data['Percent'] > 0.15].
          index, axis=1, inplace=True)
print(data.shape)
```

```python
# 对于部分类别型数据，其空值可用众数进行填充，如 Electrical
data['Electrical'].fillna(data['Electrical'].mode()[0], inplace=True)
```

```python
# 由data_description里面的内容的可知，对于部分类别型数据，
#对应的的空值表示没有该项设施，则空值可以用‘None’填充,即将空值视为新的类别
cols1 = ['GarageFinish', 'GarageQual', 'GarageType', 'GarageCond', 
         'BsmtFinType2', 'BsmtExposure',
         'BsmtFinType1', 'BsmtQual', 'BsmtCond', 'MasVnrType']
for col in cols1:
    data[col].fillna('None', inplace=True)
```

```python
# 由data_description里面的内容的可知，对于数值型数据，如MasVnrArea
#表示砖石贴面面积，如果一个房子本身没有砖石贴面，则缺失值就用0来填补
data['MasVnrArea'].fillna(0, inplace=True)
```

```python
# 对于 GarageYrBlt车库建造时间，空值表示未建造，可将其离散化，用None填充
year_map = pd.concat(pd.Series('YearGroup' + str(i+1), 
           index=range(1871+i*20,1891+i*20)) for i in range(0, 7))
data['GarageYrBlt'] = data['GarageYrBlt'].map(year_map)
data['GarageYrBlt']= data['GarageYrBlt'].fillna('None')# 必须离散化之后再对应映射
```

```python
# 对类别型数据做方差分析
def anova(frame):
    anv = pd.DataFrame()
    anv['feature'] = category
    pvals = []
    for c in category:
        samples = []
        for cls in frame[c].unique():
            s = frame[frame[c] == cls]['SalePrice'].values
            samples.append(s)
        pval = stats.f_oneway(*samples)[1]
        pvals.append(pval)
    anv['pval'] = pvals
    return anv.sort_values('pval')



a = anova(data)
a['-lg(pval)'] = np.log10(1./a['pval'].values)
plt.figure(figsize=(15,8))
sns.barplot(data=a, x='feature', y='-lg(pval)')
x=plt.xticks(rotation=90)
```

```python
# 构建模型
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import KFold, cross_val_score
import xgboost as xgb
```

```python
# 标准化
x = RobustScaler().fit_transform(df.drop(['SalePrice'], axis=1).values)
y = df['SalePrice'].values
```

```python
# 定义验证函数,使用5折交叉验证，采用均方根误差判别
def rmse_cv(model):
    kf = KFold(n_splits=5, shuffle=True, random_state=0)
    rmse= np.sqrt(-cross_val_score(model, x, y, scoring="neg_mean_squared_error", cv = kf))
    
    return rmse
```

```python
# GBDT算法
GBoost = GradientBoostingRegressor(n_estimators=3000, learning_rate=0.005,
                                   max_depth=4, max_features='sqrt',
                                   min_samples_leaf=15, min_samples_split=10, 
                                   loss='huber', random_state =5)
score1 = rmse_cv(GBoost)
print("Gradient Boosting score: {:.6f} ({:.6f})\n".format(score1.mean(), score1.std()))
```

```python
# XGBoost算法
model_xgb = xgb.XGBRegressor(colsample_bytree=0.36, gamma=0.042, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.88, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, silent=1,
                             random_state = 1, nthread = -1)
score2 = rmse_cv(model_xgb)
print("Xgboost score: {:.6f} ({:.6f})\n".format(score2.mean(), score2.std()))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
GBoost.fit(x_train, y_train)
y_hat = GBoost.predict(x_test)
print("RMSE =  %.4f" % np.sqrt(np.mean((y_hat - y_test) ** 2)))
print('R2_train = %.4f' % GBoost.score(x_train, y_train))
print('R2_test = %.4f' % GBoost.score(x_test, y_test))
```

```python
order = np.argsort(y_test, axis=0)
y_test = y_test[order].reshape(-1, 1)
y_hat = y_hat[order].reshape(-1, 1)

plt.figure(figsize=(8, 6),facecolor='w')
t = np.arange(len(y_test))
plt.plot(t, y_test, 'r-', linewidth=2, label='Original')
plt.plot(t, y_hat, 'g-', linewidth=2, label='Predicted')
plt.legend(loc='upper right')
plt.grid(b=True)
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
model_xgb.fit(x_train, y_train)
y_hat = model_xgb.predict(x_test)
print("RMSE =  %.4f" % np.sqrt(np.mean((y_hat - y_test) ** 2)))
print('R2_train = %.4f' % model_xgb.score(x_train, y_train))
print('R2_test = %.4f' % model_xgb.score(x_test, y_test))
```
