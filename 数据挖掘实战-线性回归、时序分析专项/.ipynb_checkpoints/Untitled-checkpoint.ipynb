```python
#最小二乘法的求解过程
```

```python
# standard import
import logging
import os

import numpy


# 样本x
SAMPLE_X = np.array(data['GDP(亿元)'])

# 样本y
SAMPLE_Y = np.array(data['CO2 emissions'])

# 样本数量
N_SAMPLES = SAMPLE_X.shape[0]

# 样本x的平均值
MEAN_X = sum(SAMPLE_X) / N_SAMPLES

# 样本y的平均值
MEAN_Y = sum(SAMPLE_Y) / N_SAMPLES

# 权重
WEIGHT = ((sum(SAMPLE_X[i] * SAMPLE_Y[i] for i in range(N_SAMPLES)) - N_SAMPLES * MEAN_X * MEAN_Y) /
          sum((SAMPLE_X[i] - MEAN_X) * (SAMPLE_X[i] - MEAN_X) for i in range(N_SAMPLES)))

# 偏置
BIAS = MEAN_Y - WEIGHT * MEAN_X
plt.figure(figsize=(9, 6),dpi=300)  
# 绘制样本
plt.plot(SAMPLE_X, SAMPLE_Y, 'ro', label='样本')
# 绘制拟合直线
plt.plot(SAMPLE_X, SAMPLE_X * WEIGHT + BIAS, label='拟合')
# 显示图例
plt.legend()
# 显示标题
plt.title('GDP与CO2线性关系')
# 保存图片
plt.savefig('./img/GDP与CO2线性关系.png')
# 显示图片
plt.show()
```

```python
import numpy as np
from atrader import *
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import pandas as pd
from sklearn.metrics import r2_score
set_setting('ALLOW_CONSOLE_SYSTEM_WARN', False)
import statsmodels.api as sm
```

```python
train_x = data[['GDP(亿元)','per human capital (yuan/person)','FDI（亿美元）']]
train_y = data['CO2 emissions']

#分割数据，将25%的数据作为测试集，其余作为训练集
train_x,test_x,train_y,test_y = train_test_split(train_x,train_y,test_size=0.25)
```

```python
#线性回归多因子 CO2预测
```

```python
def co2_predict():
    # 获取各因子数据
    x1 = data['GDP(亿元)']
    x2 = data['per human capital (yuan/person)']
    x3 = data['FDI（亿美元）']
    y1 = data['CO2 emissions']

    # 以图的方式展示单个因子与co2的关系
    fig = plt.figure(figsize=(20,5))
    fig1 = plt.subplot(131)
    plt.scatter(x1,y1)

    fig1 = plt.subplot(132)
    plt.scatter(x2, y1)

    fig1 = plt.subplot(133)
    plt.scatter(x3, y1)
    
    plt.savefig("./img/各因子与co2的回归关系")
    plt.show()
    
    # 创建模型
    model = LinearRegression()
    # 模型训练
    model.fit(train_x,train_y)
    # 模型预测
    predict_y = model.predict(test_x)
    #算出它的正确率
    score = r2_score(test_y, predict_y)
    print('多因子线性回归预测正确率:',score)
```

```python
co2_predict()
```
